```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
pokemon = pd.read_csv('pokedex.csv')
print(pokemon.shape)
pokemon.head(10)
```

```python
pkmn_types = pokemon.melt(id_vars = ['#', 'Name'],
                         value_vars = ['Type 1', 'Type 2'],
                         var_name= 'type_level', value_name = 'type').dropna()
# I HAVE DATA WRANGELED TO GET BOTH FIRST & SECOND TYPES ON A SINGLE COLUMN DATAFRAME
pkmn_types[976:986]
```

Now, we are plotting the bar graph using the countplot function by seaborn library.

```python
type_counts = pkmn_types['type'].value_counts()
type_order = type_counts.index # This gives the order from most common label to least common label.
```

```python
base_color = sb.color_palette()[0] # this gives the base color which is single color to all the bars.
sb.countplot(data = pkmn_types, y = 'type', color = base_color, order = type_order);

# Changing the plot to Relative frequency
plt.xticks(tick_props * n_pokemon,tick_names)
plt.xlabel('Proportion'); # This changes the x axis label to Proportions.
```
