```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('fuel_econ.csv')
print(fuel_econ.shape)
fuel_econ.head(6)
```

```python
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', x_jitter = 0.3, 
           scatter_kws = {'alpha' : 1/20});
```

We have applied some jitter on x axis of 0.3 and gave a transparency of 1/20 which is not close to zero.

0 = fully transparent, not visible 

1 = fully visible/opaque
